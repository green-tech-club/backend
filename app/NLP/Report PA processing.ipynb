```python
import pandas as pd
```

```python
# Dataframe
df = pd.DataFrame(columns=['Section','Country','Topic','Amount','Year'])
```

```python
import re
# function to preprocess speech
def clean(text):
    
    # removing paragraph numbers
    text = re.sub('[0-9]+.\t','',str(text))
    # removing new line characters
    text = re.sub('\n ','',str(text))
    text = re.sub('\n',' ',str(text))
    # removing apostrophes
    text = re.sub("'s",'',str(text))
    # removing hyphens
    text = re.sub("-",' ',str(text))
    text = re.sub("— ",'',str(text))
    # removing quotation marks
    text = re.sub('\"','',str(text))
    # removing salutations
    text = re.sub("Mr\.",'Mr',str(text))
    text = re.sub("Mrs\.",'Mrs',str(text))
    # removing any reference to outside text
    text = re.sub("[\(\[].*?[\)\]]", "", str(text))
    
    return text

df.loc[0,'Section']= clean(text)
```

```python
import re

def sentences(text):
    # split sentences and questions
    text = re.split('(?<=[.!?])\s+(?=[A-Z]|$)', text)
    clean_sent = []
    for sent in text:
        clean_sent.append(sent.strip())
    return clean_sent

# sentences
#df.loc[0,'sent'] = [sentences(df.loc[0,'Section'])]
```

```python
# to check if keyswords like 'programs','schemes', etc. present in sentences

def prog_sent(text):
    
    patterns = [r'\b(?i)'+'for'+r'\b',
               r'\b(?i)'+'to'+r'\b']

    output = []
    flag = 0
    for pat in patterns:
        if re.search(pat, text) != None:
            flag = 1
            break
    return flag 

# apply function
df2['Check_Schemes'] = df2['sentence'].apply(prog_sent)
```

```python
from spacy.matcher import Matcher

nlp = spacy.load('en_core_web_sm')

def all_schemes(text, check):
    schemes = []
    funding = {}
    doc = nlp(text)

    prog_list = ['dedicated to', 'contributed to', 'support for', 'contributions to', 'aid for']
    pattern = [{"POS": "ADP"}, {"POS" : "NOUN"}, {"POS" : "NOUN"}]
    funding_pattern = [{"TEXT": {"REGEX": "\d+(\.\d+)?\s*[€|$]"}}]
    year_pattern = [{"TEXT": {"REGEX": "20[1-2]\d"}}]

    if check == 0:
        return schemes

    matcher = Matcher(nlp.vocab)
    matcher.add("matching", [pattern])
    matches = matcher(doc)
    
    funding_matcher = Matcher(nlp.vocab)
    funding_matcher.add("funding_matching", [funding_pattern, year_pattern])

    for match_id, start, end in matches:
        noun_phrase = doc[start+1:end]
        scheme = noun_phrase.text
        schemes.append(scheme.strip())
        
    funding_matches = funding_matcher(doc)

    for funding_match_id, funding_start, funding_end in funding_matches:
            funding_value = doc[funding_start:funding_end].text.strip()
            funding_year = doc[funding_end:funding_end+1].text.strip()
            funding[(scheme.strip(), funding_year)] = funding_value

    return schemes, funding

# apply function
# df2['Schemes1'] = df2.apply(lambda x:all_schemes(x.sentence,x.Check_Schemes),axis=1)
```
